```python
# Import necessary libraries
import pandas as pd # data manipulation
import numpy as np # numerical python - linear algebra

from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# load the dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/PB_All_2000_2021.csv', sep=';')
df
```

```python
# Missing values
df.isnull().sum()
```

```python
# date is in object - date format
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
df
```

```python
df.info()
```

```python
df = df.sort_values(by=['id', 'date'])
df.head()
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
df.head()
```

```python
df.columns
```

```python
pollutants = ['O2', 'NO3', 'NO2', 'SO4', 'PO4', 'CL']
```

Week 2

```python
# Handle missing values
# Fill missing values (mean, median, forward/backward fill)
df.ffill(inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
# Feature and target selection - Feature - independent variable and Target dependent variable
X = df[['id', 'year']]
y = df[pollutants]
```

```python
# Encoding - onehotencoder - 22 stations - 1 - 1
X_encoded = pd.get_dummies(X, columns=['id'], drop_first=True)
```

```python
# Train, Test and Split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

```python
# Train the model
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
model.fit(X_train, y_train)
```

```python
# Evaluate model
y_pred = model.predict(X_test)
```

```python
print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('   MSE:', mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    print('   R2:', r2_score(y_test.iloc[:, i], y_pred[:, i]))
    print()
```

```python
station_id = '22'
year_input = 2024

input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
input_encoded = pd.get_dummies(input_data, columns=['id'])

# Align with training feature columns
missing_cols = set(X_encoded.columns) - set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col] = 0
input_encoded = input_encoded[X_encoded.columns]  # reorder columns

# Predict pollutants
predicted_pollutants = model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f"  {p}: {val:.2f}")
```

```python
import joblib

joblib.dump(model, 'pollution_model.pkl')
joblib.dump(X_encoded.columns.tolist(), "model_columns.pkl")
print('Model and cols structure are saved!')
```

```python
# Predict using the trained model
y_pred = model.predict(X_test)

# Convert to DataFrame for better visualization
y_pred_df = pd.DataFrame(y_pred, columns=y_test.columns)
y_test_df = y_test.reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot actual vs predicted for each pollutant
for col in y_test.columns:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=y_test_df[col], y=y_pred_df[col], alpha=0.6)
    plt.xlabel(f"Actual {col}")
    plt.ylabel(f"Predicted {col}")
    plt.title(f"Actual vs Predicted - {col}")
    plt.plot([y_test_df[col].min(), y_test_df[col].max()], [y_test_df[col].min(), y_test_df[col].max()], 'r--')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
